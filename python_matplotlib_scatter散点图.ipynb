```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np 
n = 1024
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)
t = np.arctan2(y,x) #每个点的颜色
plt.scatter(x,y,s=75,c=t,alpha=.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.xticks(()) #隐藏x轴
plt.yticks(()) #隐藏y轴
plt.show()
```
