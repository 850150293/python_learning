```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1,2,50)
y = x**2
plt.figure()
plt.plot(x,y)
plt.xlim((1,3))
plt.ylim((-1,4))

plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5)
plt.xticks(new_ticks)

plt.yticks([-2,-1.5,-1,2],[r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])

plt.show()
```
