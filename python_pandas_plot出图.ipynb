```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Series
data = pd.Series(np.random.randn(1000),index=np.arange(1000)) #index默认为从0开始的整数
data.cumsum()
data.plot()
plt.show()
```

```python
data = pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=list("ABCD"))
print(data.head(5))
data.cumsum()
data.plot()
plt.show()
```

```python
ax = data.plot.scatter(x='A',y='B',color='Red',label='Class 1')
data.plot.scatter(x='A',y='C',color='Yellow',label='Class 2',ax=ax)
plt.show()
```
