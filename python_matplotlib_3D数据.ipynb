```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

#åœ¨çª—å£ä¸Šæ·»åŠ 3Dåæ ‡
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow')) #strideè¡Œ åˆ—è·¨åº¦
#ax.plot_surface(X,Y,Z,rstride=5,cstride=5,cmap=plt.get_cmap('rainbow')) #strideè¡Œ åˆ—è·¨åº¦
ax.set_zlim(-2,2)
ax.contourf(X,Y,Z,zdir='y',offset=-4,cmap=plt.get_cmap('rainbow'))
plt.show()
```
