```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

#在窗口上添加3D坐标
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow')) #stride行 列跨度
#ax.plot_surface(X,Y,Z,rstride=5,cstride=5,cmap=plt.get_cmap('rainbow')) #stride行 列跨度
ax.set_zlim(-2,2)
ax.contourf(X,Y,Z,zdir='y',offset=-4,cmap=plt.get_cmap('rainbow'))
plt.show()
```
