```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,50)
y1 = x*2+1
y2 = x**2

l1, = plt.plot(x,y1,label='liner line')
l2, = plt.plot(x,y2,color='red',linewidth=1.0,linestyle='--',label='squar line')
#plt.legend(loc='upper right') legend将要显示的信息来自于上面的labeel
plt.legend(handles=[l1,l2],labels=['up','down'],loc='best')#loc='best'best自动分配最佳位置，还有其他
plt.show()
```
